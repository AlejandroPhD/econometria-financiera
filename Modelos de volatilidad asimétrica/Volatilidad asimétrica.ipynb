**1. INTRODUCCIÓN**

Los modelos de volatilidad asimétrica son un avance crucial en el campo de la econometría financiera, diseñados para capturar el fenómeno en el que los shocks negativos y positivos tienen diferentes impactos sobre la volatilidad. Los modelos de volatilidad tradicionales, como el modelo de heterocedasticidad condicional autorregresiva generalizada (GARCH) introducido por Bollerslev (1986), suponen efectos simétricos de los rendimientos pasados ​​sobre la volatilidad futura. Sin embargo, la evidencia empírica sugiere que los mercados financieros a menudo exhiben un efecto de apalancamiento, donde los rendimientos negativos (malas noticias) conducen a una mayor volatilidad que los rendimientos positivos (buenas noticias) de la misma magnitud. Esta asimetría en la volatilidad es particularmente evidente en los mercados de valores, donde los rendimientos negativos aumentan la percepción de riesgo y, en consecuencia, la volatilidad más que los rendimientos positivos. Para abordar esto, se han desarrollado varios modelos de volatilidad asimétrica.

Uno de los modelos de volatilidad asimétrica más notables es el modelo GARCH Exponencial (EGARCH) propuesto por Nelson (1991). A diferencia del modelo GARCH estándar, **EGARCH** modela el logaritmo de la varianza condicional, lo que permite asimetrías sin imponer restricciones de no negatividad a los parámetros. Otro modelo importante es el modelo **GJR-GARCH** introducido por Glosten, Jagannathan y Runkle (1993), que incorpora una función de indicador para dar cuenta de los diferentes impactos de los shocks positivos y negativos. El modelo de potencia asimétrica ARCH (APARCH) desarrollado por Ding, Granger y Engle (1993) amplía aún más la flexibilidad al permitir estimar el parámetro de potencia, capturando tanto la asimetría como los efectos variables de potencia en la volatilidad. Estos modelos proporcionan una representación más precisa de la dinámica del mercado al incorporar el efecto de apalancamiento y otras asimetrías, mejorando así el modelado y pronóstico de series de tiempo financieras (Engle, 1982; Bollerslev, 1986).

**2. LA VOLATILIDAD ASIMÉTRICA Y EL EFECTO DEL APALANCAMIENTO**

**2.1. Fenómeno de Volatilidad Asimétrica**

La volatilidad asimétrica es un fenómeno observado en los mercados financieros donde los choques negativos (malas noticias) tienden a aumentar la volatilidad futura más que los choques positivos (buenas noticias) de la misma magnitud. Esta asimetría se puede atribuir al hecho de que los inversores suelen reaccionar de manera más pronunciada a las malas noticias, lo que provoca un aumento en la percepción del riesgo y, por lo tanto, en la volatilidad. Los modelos tradicionales de volatilidad, como el modelo GARCH estándar, no capturan esta asimetría ya que asumen que los choques positivos y negativos tienen el mismo impacto en la volatilidad futura. Para abordar esta limitación, se han desarrollado varios modelos de volatilidad asimétrica, como el modelo EGARCH (Nelson, 1991), el modelo GJR-GARCH (Glosten, Jagannathan y Runkle, 1993) y el modelo APARCH (Ding, Granger y Engle, 1993).

<img src="AVP.png" alt="Drawing" style="width: 300px;"/> Fuente: https://www.investopedia.com/terms/a/assymetricvolatility.asp


Estos modelos permiten una representación más precisa de la dinámica de la volatilidad al diferenciar entre los impactos de los choques positivos y negativos. Por ejemplo, el modelo GJR-GARCH introduce una función indicadora que permite que los choques negativos tengan un mayor impacto en la varianza condicional que los choques positivos. Esta capacidad de modelar la asimetría es crucial para una gestión efectiva del riesgo y una valoración precisa de los derivados, ya que una volatilidad subestimada puede llevar a una valoración incorrecta de los instrumentos financieros y, por ende, a una mala toma de decisiones.

**2.2. Descripción del Efecto Apalancamiento (*Leverage Effect*)**

El efecto apalancamiento es una manifestación específica de la volatilidad asimétrica, donde los rendimientos negativos incrementan más la volatilidad futura que los rendimientos positivos. Este efecto fue documentado por Black (1976) y Christie (1982), quienes observaron que las caídas en los precios de las acciones aumentan la volatilidad debido a un aumento en el apalancamiento financiero. Cuando el valor de mercado de una empresa disminuye, su ratio de deuda sobre capital (apalancamiento) aumenta, lo que incrementa el riesgo percibido de la empresa y, por lo tanto, su volatilidad.

<img src="LEF.png" alt="Drawing" style="width: 300px;"/> Fuente: https://www.sciencedirect.com/science/article/abs/pii/S0304405X16301003

Empíricamente, numerosos estudios han confirmado la existencia del efecto apalancamiento en diferentes mercados y periodos de tiempo. Por ejemplo, Schwert (1989) encontró evidencia de que la volatilidad del mercado de valores estadounidense es mayor después de caídas significativas en los precios de las acciones. Estos hallazgos subrayan la importancia de incluir la asimetría en los modelos de volatilidad para reflejar adecuadamente la dinámica del mercado. Los modelos que incorporan el efecto apalancamiento, como el modelo GJR-GARCH, proporcionan una herramienta valiosa para los analistas financieros y los gestores de riesgo al permitir una mejor predicción y gestión de la volatilidad en presencia de choques negativos.

**4. ESTIMACIÓN DEL MODELO GJR**

El modelo GJR-GARCH, desarrollado por Glosten, Jagannathan y Runkle en 1993, es una extensión del modelo GARCH estándar que tiene en cuenta los efectos asimétricos de los choques positivos y negativos sobre la volatilidad. Esta asimetría, a menudo referida como el efecto apalancamiento, se observa en los mercados financieros donde los rendimientos negativos tienden a aumentar la volatilidad futura más que los rendimientos positivos de la misma magnitud. El modelo GJR-GARCH incorpora una función indicadora que diferencia entre choques positivos y negativos, capturando así el impacto distintivo de cada tipo de choque en la varianza condicional.

En términos matemáticos, el modelo GJR-GARCH(p,q) se puede expresar de la siguiente manera:

$\ 
\sigma_t^2 = \omega + \sum_{i=1}^{p} \left( \alpha_i \epsilon_{t-i}^2 + \gamma_i \epsilon_{t-i}^2 I(\epsilon_{t-i} < 0) \right) + \sum_{j=1}^{q} \beta_j \sigma_{t-j}^2 
\$

Aquí, $\sigma_t^2$ denota la varianza condicional en el tiempo $t$, $\omega$ es un término constante, $\alpha_i$ son los coeficientes para los residuos al cuadrado rezagados (términos ARCH), $\gamma_i$ son los coeficientes para los términos de asimetría, $\beta_j$ son los coeficientes para las varianzas condicionales rezagadas (términos GARCH), y $I(\epsilon_{t-i} < 0)$ es una función indicadora que toma el valor 1 si $\epsilon_{t-i}$ es negativo (es decir, el choque es negativo) y 0 en caso contrario. La inclusión del término $\gamma_i \epsilon_{t-i}^2 I(\epsilon_{t-i} < 0)$ permite que el modelo tenga en cuenta el aumento de la volatilidad asociado con los choques negativos.

En la práctica, el modelo GJR-GARCH se implementa para capturar mejor el comportamiento de los datos de series temporales financieras, particularmente al modelar activos que exhiben efectos de apalancamiento. Al diferenciar entre los impactos de los choques positivos y negativos, el modelo GJR-GARCH proporciona una representación más precisa de la dinámica de la volatilidad. Esta mayor precisión es crucial para tareas como la gestión de riesgos, la valoración de derivados y la previsión de volatilidad. La capacidad del modelo para manejar la asimetría lo convierte en una herramienta valiosa para analistas financieros e investigadores que necesitan tener en cuenta el efecto apalancamiento, a menudo observado en los mercados financieros.



Se utilizará la librería de `arch_model` para incluir la avariable dummy $\gamma_i$ a través del parámetro `o=1`.

En la biblioteca `arch` de Python, el parámetro `o=1` dentro de la función `arch_model` se utiliza para especificar el orden de los términos asimétricos en el modelo GJR-GARCH. Este parámetro es crucial para capturar los efectos de apalancamiento y la asimetría en la volatilidad.

En el contexto del modelo GJR-GARCH, el parámetro `o` representa el número de términos de asimetría que se incluyen en la especificación del modelo. Cuando se establece `o=1`, se está indicando que el modelo debe incluir un término que capture la asimetría en la varianza condicional. Este término adicional permite que el modelo diferencie entre los choques positivos y negativos, permitiendo que los choques negativos (es decir, las malas noticias) tengan un mayor impacto en la volatilidad futura que los choques positivos de la misma magnitud.

```python
# Estimación del modelo GJR-GARCH
gjr_garch = arch_model(returns, vol='Garch', p=1, q=1, o=1, rescale='bool').fit(disp='off')

# Print the model summary
print(gjr_garch.summary())
```

**4.2. GRAFICA DE LA VOLATILIDAD CONDICIONAL HISTÓRICA**

La volatilidad condicional se refiere a la variabilidad esperada de los rendimientos de un activo financiero, basada en la información disponible en un momento específico del tiempo. A diferencia de la volatilidad incondicional, que mide la variabilidad total sin considerar el tiempo, la volatilidad condicional se ajusta dinámicamente para reflejar los cambios en la incertidumbre conforme llega nueva información. Esta medida es crucial en la econometría financiera, ya que permite modelar y predecir cómo los eventos recientes y las condiciones actuales del mercado influyen en la volatilidad futura. Modelos como el GARCH (Generalized Autoregressive Conditional Heteroskedasticity) y sus variantes, como EGARCH y GJR-GARCH, son utilizados para estimar y pronosticar la volatilidad condicional, proporcionando una herramienta fundamental para la gestión de riesgos y la toma de decisiones en los mercados financieros.


```python
# Graficamos la volatilidad condicional histórica
plt.figure(figsize=(10, 6))
plt.plot(gjr_garch.conditional_volatility)
plt.title('Volatilidad condicional a partir del modelo GJR-GARCH para ' + ticker)
plt.show()
```

**4.2.1. INTERPRETACIÓN DE LA GRÁFICA DE VOLATILIDAD CONDICIONAL**

La gráfica de volatilidad condicional muestra la evolución temporal de la volatilidad estimada por el modelo GJR-GARCH. Esta volatilidad condicional es una medida de la incertidumbre o riesgo asociado con los rendimientos en cada punto del tiempo. A continuación, se presenta una explicación detallada de cómo interpretar esta gráfica.

##### Descripción de la Volatilidad Condicional

La volatilidad condicional $\sigma_t$ se calcula utilizando los parámetros estimados del modelo GJR-GARCH. Esta medida refleja la varianza condicional al tiempo $t$, es decir, la varianza de los rendimientos dados los valores pasados. La fórmula general para la varianza condicional en el modelo GJR-GARCH(1,1) es:

$
\sigma_t^2 = \omega + \alpha_1 \epsilon_{t-1}^2 + \gamma_1 \epsilon_{t-1}^2 I(\epsilon_{t-1} < 0) + \beta_1 \sigma_{t-1}^2
$

Donde:
- $\sigma_t^2$ es la varianza condicional al tiempo $t$.
- $\omega$ es el término constante.
- $\alpha_1$ es el coeficiente para los residuos al cuadrado rezagados.
- $\gamma_1$ es el coeficiente para los efectos asimétricos (choques negativos).
- $I(\epsilon_{t-1} < 0)$ es una función indicadora que toma el valor 1 si $\epsilon_{t-1}$ es negativo y 0 en caso contrario.
- $\beta_1$ es el coeficiente para las varianzas condicionales rezagadas.
- $\epsilon_{t-1}$ son los residuos al tiempo $t-1$.

##### Interpretación de la Gráfica

1. **Patrones de Volatilidad**: La gráfica de volatilidad condicional permite identificar periodos de alta y baja volatilidad. Períodos con picos pronunciados indican alta volatilidad, lo que significa que hay mayor incertidumbre o riesgo en los rendimientos durante esos periodos.

2. **Efectos de Choques Negativos**: En el modelo GJR-GARCH, los choques negativos tienen un impacto mayor en la volatilidad futura debido al término $\gamma_1 \epsilon_{t-1}^2 I(\epsilon_{t-1} < 0)$. Por lo tanto, si se observan picos de volatilidad después de grandes caídas en los rendimientos, esto es consistente con el efecto apalancamiento.

3. **Persistencia en la Volatilidad**: La presencia de autocorrelación en la volatilidad se observa si la volatilidad alta o baja tiende a persistir durante varios periodos. Esto es capturado por el coeficiente $\beta_1$, que indica cuánto de la volatilidad pasada se transmite a la volatilidad futura.

```python
# Extraemos los parámetros del modelo GJR
params = gjr_garch.params
omega = params['omega']
alpha = params['alpha[1]']
gamma = params['gamma[1]']
beta = params['beta[1]']
```

La función `np.linspace` de la biblioteca `NumPy` genera un conjunto de valores igualmente espaciados dentro de un rango especificado. En este caso, `np.linspace(-0.4, 0.4, 100)` crea un array de 100 valores que van desde -0.4 hasta 0.4. Estos valores representan diferentes niveles de shocks que pueden afectar la volatilidad condicional.

El siguiente código genera un rango de shocks y calcula cómo cada uno de estos shocks afectaría la varianza condicional de acuerdo con los parámetros de un modelo GJR-GARCH. Esto permite visualizar la relación entre los shocks de mercado y la respuesta de la volatilidad condicional, lo cual es útil para entender la dinámica del riesgo en diferentes escenarios de mercado.

```python
# Definimos un rango de shocks
shock_range = np.linspace(-0.4, 0.4, 100)

# Calculamos la varianza condicional (sigma^2) para cada shock
conditional_variance = omega + alpha * shock_range**2 + gamma * (shock_range**2) * (shock_range < 0) + beta * (0)

# Calculamos la desviación estándar (sigma)
sigma = np.sqrt(conditional_variance)
```

**4.4.2. VOLATILIDAD INCONDICIONAL**

#### Que entendemos por Volatilidad Incondicional

La volatilidad incondicional en el modelo GJR-GARCH representa la varianza promedio a largo plazo de los rendimientos financieros, incorporando tanto los efectos simétricos como asimétricos de los choques. Este modelo extiende el GARCH estándar al incluir un término adicional que captura la asimetría de los choques negativos, lo que resulta en una volatilidad más alta en respuesta a malas noticias comparado con buenas noticias de la misma magnitud. La fórmula para la varianza incondicional en el modelo GJR-GARCH es $\sigma^2 = \frac{\omega}{1 - \alpha - \frac{\gamma}{2} - \beta}$, donde $\omega$ es el término constante, $\alpha$ es el coeficiente de los residuos al cuadrado, $\gamma$ es el coeficiente de los efectos asimétricos, y $\beta$ es el coeficiente de la varianza condicional rezagada. Esta varianza incondicional proporciona una medida del nivel de riesgo inherente del activo financiero en ausencia de choques adicionales, siendo fundamental para la evaluación de la estabilidad y el riesgo a largo plazo en los mercados financieros.

En el modelo GJR-GARCH, la variable $\gamma$ captura el efecto asimétrico de los choques negativos en la volatilidad. Al calcular la varianza incondicional, se debe considerar cómo $\gamma$ contribuye al promedio general de la volatilidad a lo largo del tiempo.

Dividir $\gamma$ entre dos tiene en cuenta que, en promedio, la mitad de los choques serán negativos y la otra mitad serán positivos. Por lo tanto, el impacto promedio de $\gamma$ en la varianza a lo largo del tiempo es la mitad de su valor.

La varianza incondicional $\sigma^2$ se calcula como:

$
\sigma^2 = \frac{\omega}{1 - \alpha - \frac{\gamma}{2} - \beta}
$

Esta fórmula asegura que el impacto promedio de los choques negativos y positivos se refleje correctamente en la estimación de la varianza incondicional, proporcionando una medida precisa de la volatilidad promedio en el tiempo.


#### Estacionariedad de la volatildiad
El siguiente código verifica la condición de estacionaridad del modelo GJR-GARCH y calcula la varianza incondicional si la condición se cumple. Primero, se evalúa si la suma de los parámetros $\alpha$, $\gamma / 2$ y $\beta$ es menor que 1, lo cual es necesario para asegurar que la volatilidad sea estacionaria. Si esta condición se cumple, se calcula la varianza incondicional utilizando la fórmula $\sigma^2 = \frac{\omega}{1 - \alpha - \frac{\gamma}{2} - \beta}$ y luego se obtiene la volatilidad incondicional como la raíz cuadrada de la varianza incondicional. El resultado se imprime en la consola. Si la condición de estacionaridad no se cumple, el código imprime un mensaje indicando que el modelo no es estacionario y que la volatilidad incondicional no se puede calcular.

```python
# Asegurar estacionaridad en la volatilidad
if alpha + gamma / 2 + beta < 1:
    unconditional_variance = omega / (1 - alpha - gamma / 2 - beta)
    unconditional_volatility = np.sqrt(unconditional_variance)
    print(f'Volatilidad Incondicional (Largo-plazo): {unconditional_volatility}')
else:
    print('El modelo no es estacionario; La volatilidad incondicional no se puede calcular.')
```

**5. ESTIMACIÓN DEL MODELO EGARCH**

#### El Modelo EGARCH

El modelo EGARCH (Exponential Generalized Autoregressive Conditional Heteroskedasticity) fue propuesto por Daniel B. Nelson en 1991. Este modelo se utiliza para capturar la dinámica de la volatilidad en series temporales financieras, permitiendo modelar la asimetría en la respuesta de la volatilidad a los choques. A diferencia del modelo GARCH, el EGARCH modela el logaritmo de la varianza condicional, asegurando que la varianza sea siempre positiva sin imponer restricciones no negativas en los parámetros.

#### Ecuación del Modelo EGARCH

La ecuación general del modelo EGARCH(p,q) es:

$
\log(\sigma_t^2) = \omega + \sum_{i=1}^{q} \beta_i \log(\sigma_{t-i}^2) + \sum_{j=1}^{p} \alpha_j \left( \frac{\epsilon_{t-j}}{\sigma_{t-j}} \right) + \sum_{j=1}^{p} \gamma_j \left( \left| \frac{\epsilon_{t-j}}{\sigma_{t-j}} \right| - E \left| \frac{\epsilon_{t-j}}{\sigma_{t-j}} \right| \right)
$

#### Explicación de las Variables

- $\sigma_t^2$: Varianza condicional en el tiempo $t$.
- $\omega$: Término constante.
- $\beta_i$: Coeficientes de los logaritmos de las varianzas condicionales rezagadas.
- $\alpha_j$: Coeficientes de los términos de choques estandarizados.
- $\gamma_j$: Coeficientes de los términos de asimetría.
- $\epsilon_t$: Residuos o innovaciones en el tiempo $t$.

#### Explicación de la Variable de Asimetría

El término $\gamma_j$ en el modelo EGARCH captura la asimetría en la respuesta de la volatilidad a los choques. Este término permite que la volatilidad reaccione de manera diferente a los choques positivos y negativos. Si $\gamma_j$ es positivo, los choques negativos aumentan la volatilidad más que los choques positivos de la misma magnitud, lo que es consistente con el efecto apalancamiento observado en los mercados financieros. La interpretación de $\gamma_j$ es crucial para entender cómo las malas noticias (choques negativos) afectan la percepción del riesgo en comparación con las buenas noticias.

#### Propuesta y Evidencia Empírica

El modelo EGARCH fue propuesto por Daniel B. Nelson en su artículo de 1991 titulado "Conditional Heteroskedasticity in Asset Returns: A New Approach" publicado en *Econometrica*. Nelson propuso este modelo para abordar las limitaciones de los modelos GARCH tradicionales, especialmente en la captura de asimetrías y la imposición de restricciones no negativas en los parámetros. La evidencia empírica ha demostrado que el modelo EGARCH es efectivo para modelar la volatilidad en diversas series temporales financieras, permitiendo una mejor comprensión y previsión del comportamiento del riesgo en los mercados.

*Nelson, D. B. (1991). Conditional Heteroskedasticity in Asset Returns: A New Approach. *Econometrica*, *59*(2), 347-370.*

#### Importancia de la Simulación en el Pronóstico del Modelo EGARCH

El pronóstico de la volatilidad condicional utilizando el modelo EGARCH requiere realizar una simulación debido a la naturaleza compleja y no lineal del modelo. El modelo EGARCH incluye términos logarítmicos y componentes asimétricos que hacen que la dinámica de la volatilidad sea altamente sensible a las condiciones iniciales y a los choques de mercado. La simulación permite capturar estas dinámicas de manera más precisa al generar múltiples trayectorias posibles de la volatilidad futura, tomando en cuenta la variabilidad e incertidumbre inherentes en el proceso estocástico. Este enfoque es crucial para obtener pronósticos robustos y confiables de la volatilidad, especialmente en horizontes de tiempo más largos donde la precisión del modelo es vital para la gestión del riesgo y la toma de decisiones financieras.

#### Tipo de Simulación Utilizada en el Pronóstico EGARCH

Para el pronóstico del modelo EGARCH, se utiliza la simulación de Monte Carlo. Este método implica generar un gran número de trayectorias aleatorias para las variables de interés, en este caso, la volatilidad condicional, utilizando los parámetros estimados del modelo. Cada trayectoria es una posible realización del proceso estocástico subyacente, incorporando tanto la evolución esperada de la volatilidad como la incertidumbre asociada con los choques futuros. Al promediar estas trayectorias, se obtiene una estimación del pronóstico de la volatilidad condicional. Este enfoque permite capturar la distribución completa de los posibles resultados, proporcionando una medida más rica y detallada del riesgo futuro comparado con métodos analíticos tradicionales.

```python
# Pronóstico de la volatilidad condicional utilizando simulación Monte Carlo
horizon = 10  # Horizonte de proyección
forecasts = egarch.forecast(horizon=horizon, method='simulation')
cond_vol_forecast = np.sqrt(forecasts.variance.values[-1, :])

# Imprimir el pronóstico de volatilidad
print(f'Volatilidad condicional para los siguientes {horizon} días:')
print(cond_vol_forecast)
```
